```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("../data/cleaned_data.csv")
```

```python
df['price_diff'] = df['total_price'] - df['base_price']
df['sales_per_unit'] = df['total_price'] / df['units_sold']
```

```python
category_sales = df.groupby('segment')['sales'].sum()
```

```python
fig, ax = plt.subplots()

category_sales.plot(kind='bar', ax=ax)
ax.set_title("Revenue by Segment")

plt.savefig("../images/dashboard_preview.png")
plt.show()
```
