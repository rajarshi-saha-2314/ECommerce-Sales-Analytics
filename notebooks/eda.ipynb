```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("../data/cleaned_data.csv")
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
trend = df.groupby('order_date')['sales'].sum()

plt.figure()
trend.plot()
plt.title("Sales Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")

plt.savefig("../images/sales_trend.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

total_revenue = df['sales'].sum()
total_orders = df.shape[0]
avg_order_value = df['sales'].mean()

fig, axes = plt.subplots(1, 3, figsize=(12, 3))

# KPI 1
axes[0].text(0.5, 0.6, f"{total_revenue:,.0f}", ha='center', fontsize=16)
axes[0].text(0.5, 0.3, "Total Revenue", ha='center')
axes[0].axis('off')

# KPI 2
axes[1].text(0.5, 0.6, f"{total_orders}", ha='center', fontsize=16)
axes[1].text(0.5, 0.3, "Total Orders", ha='center')
axes[1].axis('off')

# KPI 3
axes[2].text(0.5, 0.6, f"{avg_order_value:.2f}", ha='center', fontsize=16)
axes[2].text(0.5, 0.3, "Avg Order Value", ha='center')
axes[2].axis('off')

plt.tight_layout()
plt.savefig("../images/kpi_cards.png")
plt.show()
```
